```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns

import statsmodels.api as sm
from scipy import stats
from scipy.stats import jarque_bera

plt.style.use('ggplot')
```

```python
datos = pd.read_csv('datos_finales_subtask_2.csv')
```

```python
datos.drop('id', axis=1, inplace=True)
```

```python
clustering = datos['clustering']
global_ef = datos['global_efficiency']
clos = datos['clossenes']
degre = datos['degree_centrality']
neighbor = datos['neighbor_degree']
trans = datos['transitivity']
strong = datos['number_strong_comp']
entropy = datos['entropy_centr']
clos_entropy = datos['entropy_clos']
spar = datos['sparseness']
```

```python
#DISTRIBUCIÓN CLUSTERING

mu, sigma = stats.norm.fit(clustering)

x_hat = np.linspace(min(clustering), max(clustering), num=100)
y_hat = stats.norm.pdf(x_hat, mu, sigma)


fig, ax = plt.subplots(figsize=(5,4))
ax.plot(x_hat, y_hat, linewidth=2, label='normal')
ax.hist(x=clustering, density=True, bins=30, color="#3182bd", alpha=0.5)
ax.plot(clustering, np.full_like(clustering, -0.01), '|k', markeredgewidth=1)
ax.set_title('Distribución clustering')
ax.set_xlabel('clustering')
ax.set_ylabel('Densidad de probabilidad')
ax.legend();

print('Kursotis:', stats.kurtosis(clustering))
print('Skewness:', stats.skew(clustering))

#from scipy.stats import jarque_bera

jb_stat, p_value = jarque_bera(clustering)

print(f'Estadístico de Jarque-Bera: {jb_stat}')
print(f'P-valor: {p_value}')

alpha = 0.05
if p_value > alpha:
    print(" los datos siguen una distribución normal")
else:
    print("los datos no siguen una distribución normal")
```
