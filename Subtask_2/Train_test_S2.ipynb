```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def plot_stratify(df):
    print(df['label'].value_counts(normalize=False))
    fig, ax = plt.subplots(figsize=(6,4))
    df['label'].value_counts(normalize = False).plot.bar(color=["pink","lightblue","lightgreen","peachpuff","aquamarine","skyblue"])
    plt.show()
```

```python
data=pd.read_json('./subtask_2.jsonl',lines=True)
```

```python
plot_stratify(data)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data['text'], data['label'], test_size=0.3, random_state=42,stratify=data['label'])
```

```python
X_train=pd.DataFrame(X_train)
y_train=pd.DataFrame(y_train)
train=pd.concat([X_train,y_train],axis=1)
train=pd.merge(data,train,on='text')
train=train.drop('label_x',axis=1)
train.rename(columns={'label_y':'label'},inplace=True)
train.to_csv('train_S2.csv',index=False)
```

```python
plot_stratify(train)
```

```python
X_test=pd.DataFrame(X_test)
y_test=pd.DataFrame(y_test)
test=pd.concat([X_test,y_test],axis=1)
test=pd.merge(data,test,on='text')
test=test.drop('label_x',axis=1)
test.rename(columns={'label_y':'label'},inplace=True)
test.to_csv('test_S2.csv',index=False)
```

```python
plot_stratify(test)
```
